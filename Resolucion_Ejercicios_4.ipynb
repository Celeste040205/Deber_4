```python
def f(x):
    return x**3 - 7*x**2 + 14*x - 6

def biseccion_tabla_csv(a, b, tol=1e-2):
    iteracion = 0
    print("Iter; a; b; c; f(c); b - c; f(c)*f(b)")
    while (b - a) / 2 > tol:
        iteracion += 1
        c = (a + b) / 2
        fc = f(c)
        fb = f(b)
        cumple = fc * fb <= 0
        print(f"{iteracion}; {a:.12f}; {b:.12f}; {c:.12f}; {fc:.8e}; {b - c:.12f}; {fc *fb }")
        if cumple:
            a = c
        else:
            b = c
    return (a + b) / 2

raiz = biseccion_tabla_csv(0, 1)
print(f"\nRaíz aproximada: {raiz:.12f}")
```

Ejercicio 5-a

```python
def f(x):
    return (x + 3) * (x + 1)**2 * x * (x - 1)**3 * (x - 3)

def biseccion(a, b, tol=1e-6, max_iter=1000):
    if f(a) * f(b) >= 0:
        print(f"No se puede aplicar bisección en [{a}, {b}]: f(a)*f(b) >= 0")
        return None
    for i in range(max_iter):
        c = (a + b) / 2
        fc = f(c)
        if abs(fc) < tol or (b - a)/2 < tol:
            return c
        if f(a) * fc < 0:
            b = c
        else:
            a = c
    print(f"No convergió tras {max_iter} iteraciones en [{a}, {b}]")
    return (a + b) / 2

# Ceros conocidos de la función
ceros = [-3, -1, 0, 1, 3]

# Función para imprimir resultado y respuesta a la pregunta
def evaluar_intervalo(nombre, a, b):
    raiz = biseccion(a, b)
    if raiz is not None:
        cero_cercano = min(ceros, key=lambda z: abs(z - raiz))
        print(f"\nIntervalo {nombre}) [{a}, {b}]")
        print(f"Raíz aproximada: {raiz:.6f}")
        print(f"El método de bisección converge al cero en x = {cero_cercano}")
        print(f"→ Respuesta: En el intervalo {nombre}) el método de bisección converge al cero x = {cero_cercano}")
    else:
        print(f"\nIntervalo {nombre}) [{a}, {b}] no válido para bisección.")

# Evaluaciones para los intervalos
evaluar_intervalo("a", -1.5, 2.5)   # debería converger a x = 1
```

```python
import math

# Datos
L = 10.0  # cm
r = 1.0   # cm
V_obj = 12.4  # cm^3

def volumen(h):
    # Volumen de agua con h distancia desde la parte superior
    term1 = 0.5 * math.pi * r**2
    term2 = r**2 * math.asin(h / r)  # arcsin(h/r)
    term3 = h * math.sqrt(r**2 - h**2)
    area = term1 - term2 - term3
    V = L * area
    return V

def f(h):
    return volumen(h) - V_obj

def biseccion(a, b, tol=0.01, max_iter=1000):
    if f(a) * f(b) > 0:
        print("No se puede aplicar bisección en el intervalo dado.")
        return None
    iteracion = 0
    print("Iteración; a; b; c; f(c); b - c; f(c)*f(b)")
    while (b - a) / 2 > tol and iteracion < max_iter:
        iteracion += 1
        c = (a + b) / 2
        fc = f(c)
        fb = f(b)
        print(f"{iteracion}; {a:.6f}; {b:.6f}; {c:.6f}; {fc:.6e}; {b - c:.6f}; {fc*fb:.6e}")
        if fc == 0 or (b - a)/2 < tol:
            return c
        if fc * fb < 0:
            a = c
        else:
            b = c
    return (a + b) / 2

a = 0.0
b = r

profundidad = biseccion(a, b)

if profundidad is not None:
    print(f"\nLa profundidad del agua h es aproximadamente {profundidad:.4f} cm")
else:
    print("No se encontró solución en el intervalo dado.")
```

```python
import math

# Datos
s0 = 300.0    # metros
m = 0.25      # kg
g = 9.81      # m/s^2
k = 0.1       # Ns/m

def s(t):
    term1 = s0
    term2 = (m * g / k) * t
    term3 = (m**2 * g / k**2) * (1 - math.exp(- (k/m) * t))
    return term1 - term2 + term3

def f(t):
    return s(t)  # queremos que s(t) = 0

def biseccion(a, b, tol=0.01, max_iter=1000):
    if f(a) * f(b) > 0:
        print("No se puede aplicar bisección en el intervalo dado.")
        return None
    iteracion = 0
    print("Iteración; a; b; c; f(c); b - c; f(c)*f(b)")
    while (b - a) / 2 > tol and iteracion < max_iter:
        iteracion += 1
        c = (a + b) / 2
        fc = f(c)
        fb = f(b)
        print(f"{iteracion}; {a:.6f}; {b:.6f}; {c:.6f}; {fc:.6e}; {b - c:.6f}; {fc*fb:.6e}")
        if abs(fc) < 1e-8 or (b - a)/2 < tol:
            return c
        if fc * fb < 0:
            a = c
        else:
            b = c
    return (a + b) / 2

# Elegimos un intervalo razonable para t.
# t mínimo es 0, t máximo podemos poner un valor alto, por ejemplo 60 segundos
a = 0.0
b = 60.0

tiempo = biseccion(a, b)

if tiempo is not None:
    print(f"\nEl tiempo que tarda en golpear el piso es aproximadamente {tiempo:.2f} segundos")
else:
    print("No se encontró solución en el intervalo dado.")
```

```python
import math

def f(x):
    return x**3 - x - 1

def biseccion(a, b, tol=1e-4, max_iter=100):
    if f(a) * f(b) > 0:
        print("No se puede aplicar bisección: f(a) y f(b) tienen el mismo signo.")
        return None
    iteracion = 0
    print("Iteración; a; b; c; f(c); b - c; f(c)*f(b)")
    while (b - a) / 2 > tol and iteracion < max_iter:
        iteracion += 1
        c = (a + b) / 2
        fc = f(c)
        fb = f(b)
        print(f"{iteracion}; {a:.6f}; {b:.6f}; {c:.6f}; {fc:.6e}; {b - c:.6f}; {fc*fb:.6e}")
        if abs(fc) < 1e-8 or (b - a) / 2 < tol:
            return c
        if fc * fb < 0:
            a = c
        else:
            b = c
    return (a + b) / 2

# Intervalo dado
raiz_aproximada = biseccion(1, 2, tol=1e-4)

if raiz_aproximada is not None:
    print(f"\nRaíz aproximada con precisión 1e-4: {raiz_aproximada:.6f}")
else:
    print("No se encontró raíz en el intervalo.")
```
